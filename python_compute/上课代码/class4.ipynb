```python
alist = [13, 5, [1,2]]
blist = alist # 引用赋值，浅引用，两个完全一样，blist是alist的别名
print(alist is blist) # True
blist[2].append(3)
alist[1] = 6
print('alist:',alist)
print('blist:',blist)
alist.append(['增加了','一个列表'])
print('alist:',alist)
print('blist:',blist)
clist = reversed(alist) # 迭代器，只有一次生命周期
alist.reverse()
list(clist)
print('使用过一次后的迭代器clist:',list(clist))
```

```python
# 如何进行深拷贝
import copy
'''
python中，浅拷贝只对第一层创建副本，嵌套层还是原来的引用
深拷贝会完全创立副本，需要用到copy模块
列表的浅拷贝方式[::]
'''
alist = [13, 5, 17,['one', 'list']]
dlist = copy.deepcopy(alist) # 深拷贝，dlist是alist的副本
alist.append(['增加了','一个列表'])
clist = alist.copy()
alist[1] = 1 # 修改alist第一层对clist无影响，
alist[-1].append('我在嵌套层修改了对象') # 但修改第二层会影响clist
print('clist:',clist)
print('alist:',alist)
print('dlist:',dlist)
```

```python
alist = [1, 2, 3]
blist = reversed(alist) # 返回了一个迭代器，但如果没读取blist，就还没进行反转
# x = list(blist) 
alist.reverse() # 反转列表
y = list(blist) # 这里读值了，对blist才进行反转操作，此时alist是321，所以blist是123
print(y)
# 列表内置函数的返回值是None
print(alist.reverse(), list(blist)) # reverse函数返回None,list(blist)返回[]是因为blist生命周期到头了
```

```python
'''
列表推导式：[expression for item1 in iterable1 if condition1
                      for item2 in iterable2 if condition2]
'''
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed = [[row[i] for row in matrix] for i in range(3)] # 重温复习列表推导式进行矩阵的转置
print(transposed)
# 让矩阵输出自动换行
```

```python
# 生成器推导式，没有元组推导式
g = ((i + 2) ** 2 for i in range(10))
print(g) # 返回一个生成器对象
print(tuple(g)) # 生成器对象是惰性求值的，只有在需要的时候才会计算
print(tuple(g)) # 计算到头也就没有了

g = ((i + 2) ** 2 for i in range(10))
print(g.__next__())
for i in g:
    print(i, end = ' ')
```

```python
# globals()返回当前全局（模块级别）作用域的变量字典
# locals()返回当前局部作用域（函数内部，类内部）的变量字典
a = 10
def func():
    b = 20
    print('locals:', locals())
    print('globals:', globals())
func()
print('全局变量a:',globals()['a']) # 通过globals()获取全局变量
```

```python
# 字典
key = ['a', 'b', 'c']
value = [1, 2, 3,4]
c = zip(key, value) # 是一个元组迭代器，要用for遍历，或者进行类型转换访问
for i in c:
    print(i)
d = dict(zip(value, key)) # 会取较短的一个序列组成元组，再转成了字典
print(d)
x = dict()
print(x)
d = dict(name = 'Dong', age = 37)
print(d)
adict = dict.fromkeys(['name', 'age', 'sex']) # 生成一个字典的键，值为None
print(adict)
```

```python
key = ['a', 'b', 'c']
value = [1, 2, 3]
d = dict(zip(key, value))
print(d)
print(d['a'])
print(d.get('a')) # 取出的是值
# for x in d: # 默认对象是key
#     print(x)
# for key, value in d.items():
#     print(key, value)
print(d.keys())
print(d.values())
d.update({'name': 'Dong', 'age': 37}) # 追加字典
print(d)
```

```python
import string 
import random
x = string.digits # 提取数字字符组成一个字符串
print(x,type(x))
y = [random.choice(x) for i in range(20)] # 随机选200次字符
print(y)
z = ''.join(y) # 对一个列表，使用字符串的join方法，可以将列表中的元素连接成一个字符串
print(z)
d = dict()
for ch in z:
    d[ch] = d.get(ch, 0) + 1 # get(key, 0)查找key的值，如果没有就返回0
print(d)
e = dict()
e1 = e.get('1', 0) # 如果没有就返回0
print(e1)
```
