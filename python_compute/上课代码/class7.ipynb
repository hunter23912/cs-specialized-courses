```python
import re # 四个模式匹配
# 匹配对象的5个常用方法

text = 'hello world 123456 hello world 123456'

# 字符串开头加r表示raw原始字符串，去除\的转义功能
# match从头匹配
m = re.match(r'hello', text)
if m:
    print('match:', m.group())  # 返回匹配的字符串
    
# search全文查找第一个匹配
s = re.search(r'\d+', text)
if s:
    print('search:', s.group())  # 返回匹配的字符串

# findall查找所有匹配，返回列表
lst = re.findall(r'\d+',text)
print('findall:', lst)  # 返回匹配的字符串列表

# finditer查找所有匹配，返回迭代器
it = re.finditer(r'\d+', text)
for i in it:
    print('finditer:', i.group())  # 返回匹配的字符串

# m.start()
# m.end()
# m.span()
# m.group()
```

```python
# 嵌套函数,闭包（必考）
def linear(a,b):
    def result(x):
        return a*x + b
    return result # 返回函数对象，而不是函数调用
taxes = linear(0.3, 2)
print(taxes(5))

# call方法允许对象调用自己（必考）
class linear:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def __call__(self, x):
        return self.a * x + self.b
taxes2 = linear(0.3, 2)
taxes2(5)
```

```python
# 修饰器（必考）：是函数嵌套定义的另一个重要应用，本质上也是一个函数，这个函数接收其他函数，并对其进行处理，返回一个新的函数
def before(func):
    def wrapper1(*args, **kwargs):
        print('before')
        return func(*args, **kwargs)
    return wrapper1

def after(func):
    def wrapper2(*args, **kwargs):
        result = func(*args, **kwargs)
        print('after')
        return result
    return wrapper2

@before
@after
def test():
    print('test')
test()
```

```python
# 函数嵌套之偏函数
# functools.partial：偏函数，允许我们在调用函数时，固定某些参数的值，返回一个新的函数
# 函数嵌套之柯里化
# 柯里化：将一个接受多个参数的函数转换为一个只接受一个参数的函数，并返回一个新的函数，直到所有参数都被提供
def add1(x1):
    def add2(x2):
        def add3(x3):
            return x1 + x2 + x3
        return add3
    return add2
print(add1(1)(2)(3))
```
