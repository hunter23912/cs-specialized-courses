```python
'''
map(): # 映射函数
    map(func, *iterables) -> map object
    func:函数，iterables:可迭代对象，返回一个map对象
    map()对可迭代对象的每个元素应用函数func，返回一个map对象，可迭代
filter(): # 过滤函数
    filter(func, iterable) -> filter object
    func:函数，iterable:可迭代对象，返回一个filter对象
    filter()函数用于过滤序列，过滤掉不符合条件的元素，返回一个迭代器对象。
    这两个函数都可以使用lambda表达式来简化代码。
'''
nums = [1,2,3,4]
result = list(map(lambda x: -x, nums))
print('经过map的序列：',result)

nums = [1,2,3,4]
result = list(filter(lambda x: x > 2, nums))
print('经过filter的序列：',result)
```

```python
# n位水仙花数,map()函数的使用
def func71():
    n = int(input('请输入位数:'))
    for num in range(10**(n - 1), 10**n): # 列举n位的所有数
        if sum(map(lambda i: int(i) ** n, str(num))) == num:
            print(num, end=' ')
    print()
# func71()

# n位水仙花数,filter()函数的使用
def func72():
    n = int(input('请输入位数:'))
    result = filter(
        lambda num: sum(map(lambda i: int(i) ** n, str(num))) == num, 
        range(10**(n - 1), 10**n)
    )
    for num in result:
        print(num, end=' ')
    print()
func72()
```

```python
from itertools import groupby
'''
groupby():
    groupby(iterable, key=None) -> groupby object
    iterable:可迭代对象，key:函数，返回一个groupby对象
    groupby()函数用于将相邻的重复元素分组，返回一个迭代器对象。
    该函数的key参数是一个函数，用于指定分组的依据。
count():
    count(start=0, step=1) -> count object
    start:起始值，step:步长，返回一个count对象
    count()函数用于生成一个无限的整数序列，返回一个迭代器对象。
    该函数的start参数指定起始值，step参数指定步长。
cycle():
    cycle(iterable) -> cycle object
    iterable:可迭代对象，返回一个cycle对象
    cycle()函数用于生成一个无限循环的迭代器，返回一个迭代器对象。
    该函数的参数是一个可迭代对象。
repeat():
    repeat(object, times=None) -> repeat object
    object:对象，times:次数，返回一个repeat对象
    repeat()函数用于生成一个重复的迭代器，返回一个迭代器对象。
    该函数的object参数指定要重复的对象，times参数指定重复的次数。
permutations():
    permutations(iterable, r=None) -> permutations object
    iterable:可迭代对象，r:长度，返回一个permutations对象
    permutations()函数用于生成一个排列的迭代器，返回一个迭代器对象。
    该函数的iterable参数指定要排列的对象，r参数指定排列的长度。
combinations():
    combinations(iterable, r) -> combinations object
    iterable:可迭代对象，r:长度，返回一个combinations对象
    combinations()函数用于生成一个组合的迭代器，返回一个迭代器对象。
    该函数的iterable参数指定要组合的对象，r参数指定组合的长度。
'''
a = [1,2,2,3,3,4,3,4,57,7,7,5,8,9]
a.sort()
print(groupby(a)) # 返回一个groupby对象
for(key, group) in groupby(a): # 遍历groupby对象
    print(key,len(list(group))) # key:分组的依据，group:分组的结果
```

```python
import itertools
# groupby()对相邻重复元素进行分组
a = [1,1,2,2,3,1]
for key, group in itertools.groupby(a): # key是元素的值，group是一个迭代器，含所有相同的元素
    print(key, list(group))
    
# count()生成无限递增序列
for i in itertools.islice(itertools.count(10, 2), 5): # 从10开始，每次加2，取前5个数
    print(i, end= ' ')
print()

# cycle()无限循环可迭代对象
cyc = itertools.cycle('AB')
for _ in range(4):
    print(next(cyc), end=' ')
print()

# repeat()重复对象
for x in itertools.repeat('hello',3):
    print(x, end=' ')
print()
    
# permutations()全排列
for p in itertools.permutations('abc',2):
    p = ''.join(p) # join后跟的是可迭代对象
    print(p, end=' ')
print()
    
# combinations()组合
for c in itertools.combinations('abc',2):
    c = ''.join(c)
    print(c, end=' ')
print()
```

```python
def fib(n):
    a,b = 0,1
    for _ in range(n):
        a,b = b,a+b
        if _ == 3:
            return b

def fib2(n):
    a,b=0,1
    for _ in range(n):
        yield b # 读到这里就返回b
        a,b = b,a+b

a = fib
print(a(5))

b = fib2(5)
print(next(b))
print(next(b))
print(next(b))
print(next(b))
print(next(b))
```

```python
'''
如果传递给函数的参数是可变序列，并且在函数内部对该参数进行了修改，那么在函数外部也会看到这个修改。
但重新赋值，实参不会受到影响

默认值参数

关键参数：按照参数名字传递实参，顺序不重要
'''
def f(a,b,c=5):
    print(a, b, c)
f(1, 2) # 1 2 5
```
