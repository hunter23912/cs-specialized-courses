```python
# yield用来创建生成器，返回一个迭代器对象，会记住上次执行的位置
def f():
    a,b=1,1
    while True:
        yield a,b # 当函数执行到yield时，函数会暂停执行，返回一个值给调用者，下次调用时会从上次的yield处继续执行
        a,b = b, a+b
a = f()
i = 1
while i < 10:
    print(next(a))
    i += 1 
```

```python
# 函数嵌套
def linear(a,b):
    def result(x):
        return a * x + b
    return result
a = linear(2,3)(100)
print(a)

# call方法
class linear:
    def __init__(self, a, b,):
        self.a, self.b = a,b
    def __call__(self, x):
        return self.a * x + self.b
a = linear(2,3)(100)
print(a)
```

```python
# 函数修饰器
def before(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        print('before')
        return result
    return wrapper

def mid(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        print('mid')
        return result
    return wrapper

def after(*lst):
    print(lst)
    def after1(func):
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            print('after')
            return result
        return wrapper
    return after1

@before
@mid
@after('laotie','666')
def test():
    print(3)
print('before test')
test()

def timer(func):
    def func2(*args):
        import time
        start = time.time()
        result = func(*args)
        end = time.time()
        print('time:', end-start)
        return result
    return func2

@timer
def test2(x,y):
    import time
    time.sleep(1)
    return x+y
print(test2(1,2))
```

```python
# random库的学习
import random
random_float = random.random() # 生成0-1之间的随机浮点数
print(random_float)

random_float2 = random.uniform(1,10) # 生成1-10之间的随机浮点数
print(random_float2)

random_int = random.randint(1,10) # 生成1-10之间的随机整数
print(random_int)

random_num = random.randrange(0,10,2) # 生成0-10之间的随机整数，步长为2
print(random_num)

my_list = [1,2,3,4,5]
random_choice = random.choice(my_list) # 从非空序列中随机选择一个元素
print(random_choice)

my_list = ['a','b','c','d','e']
random.shuffle(my_list) # 将序列中的元素随机打乱顺序
print(my_list)

random_sample = random.sample(my_list, 3)
print(random_sample) # 从序列中随机选择指定数量的元素，返回一个列表

random.seed(35) # 相同种子会产生相同的随机数
print(random.random())
random.seed(35)
print(random.random())
```

```python
# collections库的学习
from collections import namedtuple
Point = namedtuple('Point111', ['x', 'y']) # 创建带有命名字段的元组
p = Point(1,2)
print(p)
print(p.x, p.y)

from collections import deque 
d = deque([1,2,3])
d.append(4)
print(d)
right_element = d.pop() # 删除并返回队列右端元素
print(right_element)
left_element = d.popleft() # 删除并返回队列左端元素
print(left_element)

from collections import Counter
c = Counter('abracadabra')
print(c)
print(c.most_common(2)) # 返回出现频率最高的两个元素
c.update('aaaa') # 更新计数器，增加元素的计数
print(c)

from collections import OrderedDict
od = OrderedDict([('a', 1), ('b', 2), ('c', 3)]) # 有序字典，保持插入顺序
od.move_to_end('a', last=True) # 将指定键移动到字典的末尾或开头
print(od)

from collections import defaultdict
dd = defaultdict(int) # 默认值为int类型的字典，默认值为0
print(dd['noneexistent_key'])
dd['key'] = 10
print(dd['key'])

from collections import ChainMap
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
cm = ChainMap(dict1,dict2) # 将多个字典组合成一个字典，查找时从第一个字典开始查找
print(cm['a'])
print(cm['b'])
print(cm['c'])
```
