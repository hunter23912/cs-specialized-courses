- \_xxx 保护成员，这样的对象不能用from module import *导入
- \_\_xxx\_\_ 表示系统预定义的特殊成员，不能随意定义和增加
- \_\_ 类中的私有成员，子类不能访问，可通过 对象名.\_类名\_\_xxx访问

```python
class MyClass:
    def __init__(self):
        self.__private_var = 42  # 实际会被存储为 _MyClass__private_var

obj = MyClass()

# 通过改写后的名称访问
print(obj._MyClass__private_var)  # 输出: 42
```

- 类中定义的方法有四类：公有方法、私有方法、静态方法、类方法
    - 静态方法@staticmethod，类方法@classmethod，否则python默认为普通方法。
    - 静态方法不关心类和实例。类方法能访问和修改类属性。

```python
class Root:
    __total = 0
    def __init__(self, v):
        self.__value = v
        Root.__total += 1
    
    def show(self):
        print('self.__value:',self.__value)
        print('Root.__total:',Root.__total)
        
    @classmethod # 类方法
    def classShowTotal(cls):
        print(cls.__total)
    
    @staticmethod # 静态方法
    def staticShowTotal():
        print(Root.__total)
# 示例
r = Root(3)

r.show()
Root.show(r)
```

```python
class Test:
    def __init__(self, value):
        self.__value = value

    @property # 将方法转为属性，且只读，不允许修改
    def value(self):
        return self.__value

t = Test(3)
print(t.value)
t.v = 10 # 可以对实例动态添加新属性
print(t.v)
del t.v # 可删除，但value不可删除
```

```python
class Test:
    def __init__(self, value):
        self.__value = value
    def __get(self):
        return self.__value
    def __set(self, v):
        self.__value = v
    def __del(self):
        del self.__value
    value = property(__get, __set, __del) # 修饰器，定义可读可写可删除属性
    
    def show(self):
        print(self.__value)
t = Test(3)
print(t.value)
t.value = 5
print(t.value)
t.show()
del t.value
```

```python
# 函数生成器
def gen():
    cnt = 1
    while True:
        yield cnt
        cnt += 1
g = gen()
print(next(g))
print(g.__next__())

# 函数嵌套
def linear(a,b):
    def result(x):
        return a * x + b
    return result
f = linear(2,3)
print(f(5))
```

```python
# 函数修饰器
def before(func):
    def wrapper(*args, **kwargs):
        print('before function called.')
        return func(*args, **kwargs)
    print('all done.') 
    return wrapper

def after(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        print('after function called.')
        return result
    print('inner done.')    
    return wrapper

@before
@after
def test():
    print('test function called.')

print('装饰器已经定义好了，还没执行函数')
a = test()

```

```python
# 静态方法和类方法的区别演示

class Demo:
    class_var = 0

    def __init__(self):
        Demo.class_var += 1

    @staticmethod
    def static_func():
        print("静态方法：不能访问类属性和实例属性")
        # print(self)  # 实例方法不能访问
        # print(cls)   # 类方法不能访问

    @classmethod
    def class_func(cls):
        print("类方法：可以访问和修改类属性")
        print("当前类:", cls)
        print("类属性class_var:", cls.class_var)
        cls.class_var += 10

d1 = Demo()
d2 = Demo()

Demo.static_func()   # 只能做普通功能，不能访问类和实例的数据
Demo.class_func()    # 可以访问和修改类属性

print("类属性class_var现在是：", Demo.class_var)

```
